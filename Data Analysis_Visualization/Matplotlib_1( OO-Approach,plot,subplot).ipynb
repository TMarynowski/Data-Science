````python
import math as mat
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-bright')
````

### Example 1
 

````python
import numpy as np
x = np.linspace(0, 10, 20)
y = np.power(x,2)
````

````python
x
````

````python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.show()
````

## Example 3  Matplotlib Object Oriented Approach

````python
x = np.linspace(-np.pi, np.pi, 300)
y = np.sin(x)
z= np.cos(x)
````

````python
fig = plt.figure(figsize=(13,4)) #creature figure, empty canvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # add set of axes to figure: left, bottom, width, height (range 0 to 1)
axes.plot(x, y, color='blue', linewidth=3, linestyle ='--') # draw on that set of axes
axes.plot(x,z,color ='green', linewidth=3, linestyle='-.')
axes.set_xlabel('X') 
axes.set_xlim(-2*np.pi,2*np.pi)
axes.set_ylabel('Y')
axes.set_title('$\sin(x)$ vs $\cos(x)$')
axes.legend(['$\sin(x)$','$\cos(x)$'])
````
